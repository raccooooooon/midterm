```python
health = pd.read_csv("/content/drive/My Drive/Colab Notebooks/BIGMINING/health_data.csv")
```

```python
display(health.head(10))
print(health.info())
```

```python
health['기준년도'] = health['기준년도'].astype(str)

health['가입자일련번호'] = health['가입자일련번호'].astype(str)

health['시도코드'] = health['시도코드'].astype(str)  

health['구강검진수검여부'] = health['구강검진수검여부'].astype(str)

health['치아우식증유무'] = health['치아우식증유무'].astype(str)  

health['치석'] = health['치석'].astype(str)  
```

```python
print(health.describe()) 
```

```python
sns.countplot(data=health, x='흡연상태', hue='성별코드');
plt.ylabel('해당 명수')
plt.title('성별과 흡연의 관계');
```

###성별코드에서 M은 남성을, F는 여성을 의미한다.
###흡연상태의 빈도를 보아, 여성보다 남성이 흡연을 하는 경우가 많으며 흡연을 하다 금연을 한 경우도 남성의 빈도가 높음을 알 수 있다. 

```python
sns.scatterplot(data=health, x='총콜레스테롤', y='HDL콜레스테롤');
plt.title('총콜레스테롤과 HDL 콜레스테롤');
```

### HDL콜레스테롤은 작은 입자의 콜레스테롤로 세포에 이끌려간 콜레스테롤을 간으로 돌려주고 혈관 벽에 쌓인 나쁜 콜레스테롤을 없애는 역할을 하는 성분이다.
### 총콜레스테롤이 늘어남에 따라 HDL 콜레스테롤이 늘어나는 경향을 가진다고 보기 어렵다.

```python
sns.scatterplot(data=health, x='총콜레스테롤', y='LDL콜레스테롤');
plt.title('총콜레스테롤과 LDL 콜레스테롤');
```

###LDL 콜레스테롤은 입자가 매우 큰 콜레스테롤로 양이 과도하게 증가할 경우,  혈관 벽에 쌓여서 동맥경화나 각종 질병을 야기 하는 성분이다.
###그래프를 통해 총콜레스테롤이 늘어나면 LDL 콜레스테롤도 늘어남을 알 수 있다. 

```python
sns.jointplot(data=health, x="수축기혈압", y="이완기혈압", hue="흡연상태");
```

### 비흡연자의 경우 수축기혈압과 이완기혈압 모두 흡연자, 흡연경험자에 비해 높게 나타났다.

```python
sns.jointplot(data=health, x="수축기혈압", y="이완기혈압", hue="음주여부");
```

### 음주자의 경우 수축기혈압과 이완기혈압 모두 비음주자에 비해 약간 높게 나타났다.
